## MODEL TRANING BY DEFAULT HYPERPARAMETER

# Train-Test Split & Preprocessing:


```python
X=df.drop('Outcome',axis=1)
y=df['Outcome']
X
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

**standertization**

```python

from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
```

```python
X_train=scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)
```

```python
X_train
```

```python
X_test
```

# Training the Decision Tree:

```python
#First Train the model by default hyperparameter
from sklearn.tree import DecisionTreeClassifier,plot_tree
```

```python
clf=DecisionTreeClassifier()
```

```python
clf.fit(X_train,y_train)
```

```python
y_pred=clf.predict(X_test)
```

```python
(y_pred)
```

```python
#comapresion table
df_table=pd.DataFrame(data={'Acutal':y_test,'Predicated':y_pred})
df_table
```

# Model Evaluation:

```python
#Model Evaluation
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
##accuracy
accuracy=accuracy_score(y_test,y_pred)
print(accuracy)
```

```python
#classification report
print(classification_report(y_test,y_pred))
```

```python
#confusion metrics
print(confusion_matrix(y_test,y_pred))
```

## Visualizing the Tree:

```python
plt.figure(figsize=(20, 10))
plot_tree(clf,
          filled=True,         # Fill nodes with colors
          rounded=True,        # Rounded node boxes
          feature_names=X.columns,  # Show feature names
          class_names=[str(c) for c in clf.classes_],  # Show class labels
          fontsize=10)
plt.show()
```

## FINDGING BEST HYPER PARMETER

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {
    'criterion': ['gini', 'entropy', 'log_loss'],  # Splitting criteria
    'splitter': ['best', 'random'],  # How the split is chosen
    'max_depth': [None, 2, 4, 6, 8, 10, 15, 20],  # Tree depth
    'min_samples_split': [2, 5, 10, 15, 20],  # Min samples to split a node
    'min_samples_leaf': [1, 2, 4, 6, 8, 10],  # Min samples at leaf node
    'max_features': [None, 'sqrt', 'log2'],  # Number of features to consider at split
    'max_leaf_nodes': [None, 5, 10, 20, 50],  # Max number of leaf nodes
}
```

```python
grid_model=GridSearchCV(clf,param_grid=param_grid)
```

```python
grid_model.fit(X_train,y_train)
```

```python
y_pred_final=grid_model.predict(X_test)
```

```python
y_pred_final
```

```python
grid_model.best_params_
```

```python
grid_model.best_score_
```

#Final Model Evaluation:

```python

accuracy=accuracy_score(y_test,y_pred_final)
print(accuracy)
```

```python
print(classification_report(y_test,y_pred_final))
```

```python
#confusion metrics
print(confusion_matrix(y_test,y_pred))
```
