```python
img = plt.imread('/content/drive/MyDrive/Colab Notebooks/ML WORK/K-MEANS-CLUSTERING UNSUPERVISED LEARNING/kingfisher.jpeg')
print(f'The image is {img.shape[0]}px by {img.shape[1]}px')
plt.axis('off')
plt.imshow(img)
plt.show()
```

```python
X = img.reshape(-1, 3)
wcss=[]
for i in range(1,11):
    kmeans=KMeans(n_clusters=i,init='k-means++',random_state=42)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
```

```python
plt.plot(range(1,11),wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```
