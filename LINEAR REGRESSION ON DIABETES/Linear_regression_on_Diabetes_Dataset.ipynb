```python
from sklearn.datasets import load_diabetes
import pandas as pd
data_diabt=load_diabetes()
X=pd.DataFrame(data_diabt.data,columns=data_diabt.feature_names)
y=pd.Series(data_diabt.target,name="Target")
print(X.head())
print(y.head())
```

```python
#concatationation
df=pd.concat([X,y],axis=1)
df.head()
```

```python
#importing libraryis
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error,mean_squared_error
```

```python
print(X.head())
print(y.head())
```

```python
#train test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
print(X_train.info())
```

```python
#model traning
model=LinearRegression()
model.fit(X_train,y_train)
```

```python
#model predicatio
predict=model.predict(X_test)
predict
```

```python
#model evaluation
performance=mean_absolute_error(y_test,predict)
performance
```

```python
#train test split
#train test split
X_train, X_test, y_train, y_test = train_test_split(X2, y2, test_size=0.33, random_state=42)
```

```python
#model traning
model2=LinearRegression()
model2.fit(X_train,y_train)
```

```python
#model evaluation
performance=mean_absolute_error(y_test,predict)
performance
```

```python
mse=mean_squared_error(y_test,predict)
mse
```

```python
# we will check the accuracy

print('coefficient', model2.coef_)
print('intercept', model2.intercept_)
```

```python
result = pd.DataFrame({'Actual': y_test, 'Predict' : predict})
result
```

```python
#Residual plot
# Predict and compute residuals
residuals = y_test - predict
residuals
```

```python
# Residual Plot
plt.figure(figsize=(8, 5))
sns.scatterplot(x=predict, y=residuals)
plt.axhline(0, color='red', linestyle='--')
plt.xlabel("Predicted values")
plt.ylabel("Residuals")
plt.title("Residual Plot")
plt.show()
```
