IMPORTING IMPORTANT LIBRARIES

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#IMPORTING DATA
df=pd.read_csv('Advertising.csv')
df.head()
```

```python
#Removing the unwanted column
df=df.drop('Unnamed: 0',axis=1)
df.head()
```

```python
X=df.drop('Sales',axis=1)
y=df['Sales']
print(X.head(2))
print(y)
```

CONVERSION TO POLYNOMIAL FEATURES

```python
from sklearn.preprocessing import PolynomialFeatures
poly_conveter=PolynomialFeatures(degree=3,include_bias=False)
poly_feature=poly_conveter.fit_transform(X)
poly_feature
```

```python
#shape of polyfeature
poly_feature.shape
```

```python
#Now to train test split
from sklearn.model_selection import train_test_split
```

```python
 X_train, X_test, y_train, y_test = train_test_split(poly_feature, y, test_size=0.33, random_state=101)
```

X_train.shape

Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(X_train)
X_train=scaler.transform(X_train)
X_test=scaler.transform(X_test)
```

```python
X_train
```

```python
X_test
```

```python
X_train.std()#standertization
```

**Ridge Regression**


```python
#I am importing simple Ridge Regression 
from sklearn.linear_model import Ridge
model1=Ridge(alpha=1.0)
model1.fit(X_train,y_train)
```

```python
#y_test predication
y_test_predication=model1.predict(X_test)
y_test_predication
```

**MODEL EVALUATION**

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error
```

```python
MAE_Ridge=mean_absolute_error(y_test,y_test_predication)
MAE_Ridge
```

```python
RMAE_Ridge=np.sqrt(mean_squared_error(y_test,y_test_predication))
RMAE_Ridge
```

**CHOSSING BEST ALPHA VALUE**

we will need to do use cross validation to test a varity 
of alpha parameter on cross validation

```python
from sklearn.linear_model import RidgeCV
model2=RidgeCV(alphas=(0.1, 1.0, 10.0))
model2.fit(X_train,y_train)
```

```python
#selected alpha value
model2.alpha_
```

```python
#model y test predication
y_test_predication_ridgeCV=model2.predict(X_test)
y_test_predication_ridgeCV
```

```python
#parameter of x
model2.coef_
```

```python
#Model Evaluation
MAE_RidgeCV=mean_absolute_error(y_test,y_test_predication_ridgeCV)
MAE_RidgeCV
```

```python
RMAE_RidgeCV=np.sqrt(mean_squared_error(y_test,y_test_predication_ridgeCV))
RMAE_RidgeCV
```
